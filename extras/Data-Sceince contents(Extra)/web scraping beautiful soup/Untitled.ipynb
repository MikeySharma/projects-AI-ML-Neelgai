```python
from bs4 import BeautifulSoup
import requests

website = 'https://www.lyrics.com/artist/Cat-Stevens/5528'
#yesle page fetch garxa
result= requests.get(website)
#yesle page ko content lyaunxa
content = result.text
print(content)

```

```python
#create the soup
soup= BeautifulSoup(content,'lxml')
print(soup)
#we always do this regardless of the website
```

# find attribute 

### soup.find('tag', attributeName="Value") SYNTAX
### find vs findall tag

```python
#to find the titles

titles= soup.find_all('h3',class_='artist-album-label')
print(titles)

```

```python
for title in titles:
    print(title.get_text())
```
